```python
# 위치 수정 필요, 정보 추가
def get_hotels(city_id, checkin, checkout):
    # URL을 동적으로 생성
    url = f'https://kr.trip.com/hotels/list?city={city_id}&checkin={checkin}&checkout={checkout}'
    
    # 헤더 설정
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    
    # HTTP GET 요청
    response = requests.get(url, headers=headers)
    
    # 요청이 성공적으로 이루어졌는지 확인
    if response.status_code == 200:
        html = response.text
        soup = BeautifulSoup(html, 'html.parser')
        
        # 3. 호텔 정보를 담고 있는 컨테이너를 선택
        hotel_list = soup.select('div.compressmeta-hotel-wrap-v8')
    
        # 4. 호텔 정보 추출
        for hotel in hotel_list:
            # 호텔 이름 추출
            name_tag = hotel.select_one('div.list-card-title .name')
            name = name_tag.text.strip() if name_tag else "이름 정보 없음"
        
            # 이미지 URL 추출
            img_tag = hotel.select_one('img.m-lazyImg__img')
            img_url = img_tag['src'] if img_tag else "이미지 없음"
        
            # 평점 추출
            rating_tag = hotel.select_one('div.score .real')
            rating = rating_tag.text.strip() if rating_tag else "평점 없음"
        
            # 평가 추출
            rating2_tag = hotel.select_one('div.describe .outstanding')
            rating2 = rating2_tag.text.strip() if rating_tag else "평가 없음"
        
            # 리뷰 수 추출
            review_tag = hotel.select_one('div.count a')
            review = review_tag.text.strip() if review_tag else "리뷰 없음"
        
            # 코멘트 추출
            comment_tag = hotel.select_one('div.list-encourage-review .review-item')
            comment = comment_tag.text.strip() if comment_tag else "리뷰 없음"
        
            # 위치 추출
            location_tag = hotel.select_one('div.list-card-transport-v8 .transport')
            location = location_tag.text.strip() if location_tag else "위치 정보 없음"
        
            # 가격 추출
            price_tag = hotel.select_one('div.real div')
            price = price_tag.text.strip() if price_tag else "가격 정보 없음"
    
        
            # 결과 출력
            print(f"Hotel Name: {name}")
            print(f"Image URL: {img_url}")
            print(f"Rating: {rating}")
            print(f"Rating2: {rating2}")
            print(f"Reviews: {review}")
            print(f"Comment: {comment}")
            print(f"Location: {location}")
            print(f"Price: {price}")
            print('-' * 80)

    else:
        print(f"Failed to retrieve page, status code: {response.status_code}")
```

```python
# 하나의 숙소에 대한 정보 (가장 위 box)

# 웹페이지 URL
url = 'https://kr.trip.com/hotels/seoul-hotel-detail-988402/banyan-tree-club-and-spa-seoul/'

# 웹페이지 요청
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')


# 호텔 정보 추출
def extract_hotel_info(soup):
    hotel_info = {}

    # 호텔 이름
    hotel_name = soup.find('h1', class_='headInit_headInit-title_name__xKWeS')
    hotel_info['name'] = hotel_name.get_text(strip=True) if hotel_name else '정보 없음'

    # 호텔 주소
    hotel_address = soup.find('span', class_='headInit_headInit-address_text__D_Atv')
    hotel_info['address'] = hotel_address.get_text(strip=True) if hotel_address else '정보 없음'

    # 호텔 설명
    hotel_description = soup.find('div', class_='headInit_headInit-address_desc__tJfYJ')
    hotel_info['description'] = hotel_description.get_text(strip=True) if hotel_description else '정보 없음'

    # 호텔 등급
    hotel_level = soup.find('span', class_='hotelLevel_hotelLevel__mhh3v')
    stars = len(hotel_level.find_all('i')) if hotel_level else 0
    hotel_info['stars'] = stars

    # 호텔 위치
    hotel_location = soup.find_all('div', class_='headTraffic_headTraffic-item_container__c2jtJ')
    locations = []
    for item in hotel_location:
        description = item.find('div', class_='headTraffic_headTraffic-item_desc__HT0G0')
        distance = item.find('div', class_='headTraffic_headTraffic-item_distance__yCw1l')
        locations.append({
            'description': description.get_text(strip=True) if description else '정보 없음',
            'distance': distance.get_text(strip=True) if distance else '정보 없음'
        })
    hotel_info['location'] = locations

    # 호텔 시설
    hotel_facilities = soup.find_all('span', class_='headFacility_headFacility-list_desc__S8UuA')
    facilities = [facility.get_text(strip=True) for facility in hotel_facilities]
    hotel_info['facilities'] = facilities

    # 리뷰 점수 및 내용
    review_score = soup.find('b', class_='headReview_headReviewV2-scoreBox_bold__iELuw')
    review_content = soup.find('span', class_='headReview_headReviewV2-contentText__WngP8')
    hotel_info['review_score'] = review_score.get_text(strip=True) if review_score else '정보 없음'
    hotel_info['review_content'] = review_content.get_text(strip=True) if review_content else '정보 없음'

    return hotel_info

# 호텔 정보 출력
hotel_info = extract_hotel_info(soup)

# 호텔 정보 보기 좋은 형식으로 출력하기
print("호텔 정보")
print("=" * 40)
print(f"호텔 이름: {hotel_info['name']}")
print(f"주소: {hotel_info['address']}")
print(f"설명: {hotel_info['description']}")
print(f"별점: {'★' * hotel_info['stars'] if hotel_info['stars'] else '정보 없음'}")
print("\n위치 정보:")
for loc in hotel_info['location']:
    print(f" - {loc['description']}: {loc['distance']}")
print("\n시설:")
for facility in hotel_info['facilities']:
    print(f" - {facility}")
print(f"\n리뷰 점수: {hotel_info['review_score']}")
print(f"리뷰 내용: {hotel_info['review_content']}")
```

```python
# 한 호텔에 대한 리뷰(카테고리별)

# 크롤링할 URL
url = 'https://kr.trip.com/hotels/seoul-hotel-detail-988402/banyan-tree-club-and-spa-seoul/'

# 웹 페이지 요청
response = requests.get(url)
response.raise_for_status()  # 요청 실패 시 예외 발생

# HTML 파싱
soup = BeautifulSoup(response.text, 'html.parser')

# 리뷰 정보 추출
def extract_review_info(soup):
    review_info = {}
    
    # 전체 점수
    current_score = soup.select_one('.reviewScores_reviewOverallScores-currentScore__v5Qtj').text.strip()
    review_info['Overall Score'] = current_score
    
    # 점수 설명
    score_desc = soup.select_one('.reviewScores_reviewOverallScores-scoreDesc__fEv8O').text.strip()
    review_info['Score Description'] = score_desc
    
    # 검증된 리뷰 아이콘 확인
    review_info['Verified'] = '검증된 리뷰' in soup.select_one('.reviewScores_reviewOverallScores-reviewFromTrip__mlpqP').text.strip()
    
    # 카테고리별 점수
    categories = soup.select('.reviewScores_reviewCategoryScores-item__GpD9u')
    category_scores = {}
    for category in categories:
        title = category.select_one('.reviewScores_reviewCategoryScores-itemHead__4HXHu').text.strip().split()[0]
        score = category.select_one('.reviewScores_reviewCategoryScores-titleScore__pXwjH span').text.strip()
        category_scores[title] = score
    
    review_info['Category Scores'] = category_scores
    
    return review_info

# 리뷰 정보 출력
review_info = extract_review_info(soup)

# 리뷰 정보 보기 좋은 형식으로 출력하기
print("리뷰 정보")
print("=" * 40)
print(f"전체 점수: {review_info['Overall Score']}")
print(f"점수 설명: {review_info['Score Description']}")
print(f"검증된 리뷰: {'예' if review_info['Verified'] else '아니오'}")
print("\n카테고리별 점수:")
for category, score in review_info['Category Scores'].items():
    print(f" - {category}")
```

```python
# 한 호텔의 한 객실에 대한 정보 - 이미지 수정
# 웹 페이지 요청
response = requests.get(url)
response.raise_for_status()  # 요청 실패 시 예외 발생

# HTML 파싱
soup = BeautifulSoup(response.text, 'html.parser')

def extract_room_info(soup):
    room_info = {}
    
    # 객실 이미지 URL
    img_tag = soup.select_one('.spiderRoomList_baseRoomPic__oIvSX')
    room_info['Image URL'] = img_tag['src'] if img_tag else None
    
    # 객실 이름
    name_tag = soup.select_one('.spiderRoomList_roomName__5jbly')
    room_info['Room Name'] = name_tag.text.strip() if name_tag else None
    
    # 객실 면적 및 층
    facility_elements = soup.select('.spiderRoomList_roomFacility_hotFacility_element__xp02v')
    size_and_floor_tag = facility_elements[0].select_one('.spiderRoomList_iconText__67QOv') if len(facility_elements) > 0 else None
    room_info['Size and Floor'] = size_and_floor_tag.text.strip() if size_and_floor_tag else None

    # 객실 시설
    facilities = {}
    facility_list = soup.select('.spiderRoomList_roomFacility_hotFacility_element__xp02v')
    for facility in facility_list:
        icon_tag = facility.select_one('.spiderRoomList_iconText__67QOv')
        if icon_tag:
            icon_name = icon_tag.text.strip()
            facilities[icon_name] = True

    room_info['Facilities'] = facilities

    return room_info

# 객실 정보 추출
room_info = extract_room_info(soup)

# 객실 정보 보기 좋은 형식으로 출력하기
print("객실 정보")
print("=" * 40)
print(f"이미지 URL: {room_info['Image URL']}")
print(f"객실 이름: {room_info['Room Name']}")
print(f"면적 및 층: {room_info['Size and Floor']}")
print("\n시설:")
if room_info['Facilities']:
    for facility in room_info['Facilities']:
        print(f" - {facility}")
else:
    print("시설 정보 없음")
```

```python
# 한 호텔 주변에 대한 정보 -> 실행이 안 됨

# BeautifulSoup 객체 생성
soup = BeautifulSoup(html, 'html.parser')

# 교통 정보 섹션 파싱
def parse_transport_info(soup):
    transport_info = []
    transport_section = soup.find_all('div', class_='hotelLocationNew_hotelLocation-item_places__s9Jhg')
    
    for section in transport_section:
        type_span = section.find('span', class_='hotelLocationNew_highlight__GLoxa')
        if type_span:
            type_info = type_span.text
            name = section.find('span', class_='hotelLocationNew_hotelLocation-details_name__1bZeV').text.split(': ')[1]
            distance = section.find('div', class_='hotelLocationNew_hotelLocation-details_distance__3iJuS').text
            transport_info.append(f"{type_info}: {name} ({distance})")
    
    return transport_info

# 쇼핑 정보 섹션 파싱
def parse_shopping_info(soup):
    shopping_info = []
    shopping_section = soup.find('div', class_='hotelLocationNew_hotelLocation-item_shoppingPlaces__L_DOi')
    
    if shopping_section:
        shopping_items = shopping_section.find_all('div', class_='hotelLocationNew_hotelLocation-details___w7lm')
        
        for item in shopping_items:
            name = item.find('div', class_='hotelLocationNew_hotelLocation-details_name__1bZeV').text
            distance = item.find('div', class_='hotelLocationNew_hotelLocation-details_distance__3iJuS').text
            shopping_info.append(f"{name} ({distance})")
    
    return shopping_info

# 출력
def pretty_print_info(transport_info, shopping_info):
    print("교통 정보:")
    for info in transport_info:
        print(f"  - {info}")
    
    print("\n쇼핑:")
    for info in shopping_info:
        print(f"  - {info}")

# 파싱 및 출력 실행
transport_info = parse_transport_info(soup)
shopping_info = parse_shopping_info(soup)
pretty_print_info(transport_info, shopping_info)
```
