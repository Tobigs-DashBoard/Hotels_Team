```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains

# Chrome 옵션 설정
chrome_options = Options()
chrome_options.binary_location = "/Applications/Google Chrome.app/Contents/MacOS/Google Chrome"
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
chrome_options.add_experimental_option('useAutomationExtension', False)

# User-Agent 변경
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.5112.81 Safari/537.36")

# ChromeDriver 서비스 설정
service = Service("/Users/baengminji/Desktop/minji/tobigs/visualization/chromedriver-mac-arm64/chromedriver")

# WebDriver 초기화
driver = webdriver.Chrome(service=service, options=chrome_options)

# 웹 페이지 열기
driver.get("https://www.agoda.com/ko-kr/?site_id=1922887&tag=f7739694-dbb7-41bd-aa27-be7c942ce354&gad_source=1&gbraid=0AAAAAD-GdVnN8X8xnT7LkfxoClUCQ1ZJk&device=c&network=g&adid=695788229412&rand=17559854526313289191&expid=&adpos=&aud=kwd-6927948326&gclid=Cj0KCQjwiOy1BhDCARIsADGvQnAeNogLQ9nk4EqmOZ36vrRk2odOxk8q45gvTZbVE6s4CujvM75al7oaAulREALw_wcB&pslc=1&ds=XDL8X%2Bh6flWPPqAi")

# 페이지 로딩 대기 (추가 대기 시간)
time.sleep(5)



try:
    close_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//button[contains(text(),"닫기")]'))
    )
    actions = ActionChains(driver)
    actions.move_to_element(close_button).click().perform()
    print("닫기 버튼을 클릭했습니다.")
except Exception as e:
    print("닫기 버튼을 클릭하는 데 실패했습니다:", e)

# 위치 입력
try:
    location_input = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="textInput"]'))
    )
    location_input.click()  # 위치 입력란 클릭
    location_input.send_keys('서울특별시')
    print("위치 입력 완료")

    # 자동완성 목록의 첫 번째 항목 선택
    first_suggestion = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//*[@data-selenium="topDestinationListItem"]'))
    )
    first_suggestion.click()
    print("자동완성 항목 클릭 완료")
except Exception as e:
    print("위치 입력 중 오류 발생:", e)

# 날짜 선택
try:
    start_date = "2024-08-18"
    end_date = "2024-08-24"

    start_date_element = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, f'//span[@data-selenium-date="{start_date}"]'))
    )
    start_date_element.click()
    print(f"시작 날짜 {start_date} 선택 완료")

    end_date_element = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, f'//span[@data-selenium-date="{end_date}"]'))
    )
    end_date_element.click()
    print(f"종료 날짜 {end_date} 선택 완료")
except Exception as e:
    print("날짜 선택 중 오류 발생:", e)

# 객실 수 조정
try:
    current_room_count_element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//div[@data-selenium="desktop-occ-room-value"]/p'))
    )
    current_room_count = int(current_room_count_element.text)

    target_room_count = 3  # 원하는 객실 수

    while current_room_count != target_room_count:
        if current_room_count < target_room_count:
            plus_button = driver.find_element(By.XPATH, '//button[@aria-label="추가하기"]')
            plus_button.click()
            current_room_count += 1
            time.sleep(2)
        elif current_room_count > target_room_count:
            minus_button = driver.find_element(By.XPATH, '//button[@aria-label="빼기"]')
            minus_button.click()
            current_room_count -= 1
            time.sleep(2)

    final_room_count = int(driver.find_element(By.XPATH, '//div[@data-selenium="desktop-occ-room-value"]/p').text)
    print(f"설정된 객실 수: {final_room_count}")
except Exception as e:
    print("객실 수 조정 중 오류 발생:", e)

# 성인 수 조정
try:
    current_adult_count_element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//div[@data-selenium="desktop-occ-adult-value"]/p'))
    )
    current_adult_count = int(current_adult_count_element.text)

    target_adult_count = 3  # 원하는 성인 수

    while current_adult_count != target_adult_count:
        if current_adult_count < target_adult_count:
            plus_button = driver.find_element(By.XPATH, '//button[@aria-label="추가하기"]')
            plus_button.click()
            current_adult_count += 1
            time.sleep(2)
        elif current_adult_count > target_adult_count:
            minus_button = driver.find_element(By.XPATH, '//button[@aria-label="빼기"]')
            minus_button.click()
            current_adult_count -= 1
            time.sleep(2)

    final_adult_count = int(driver.find_element(By.XPATH, '//div[@data-selenium="desktop-occ-adult-value"]/p').text)
    print(f"설정된 성인 수: {final_adult_count}")
except Exception as e:
    print("성인 수 조정 중 오류 발생:", e)


try:
    close_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//button[@data-selenium="searchButton"]'))
    )
    actions = ActionChains(driver)
    actions.move_to_element(close_button).click().perform()
    print("검색 버튼을 클릭했습니다.")
    # 검색 결과 페이지 로딩 대기
    WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="Tabs-Container"]/button/div'))  # 실제 결과 섹션의 XPATH를 사용해야 함
    )
    print("검색 결과 페이지 로드 완료")
except Exception as e:
    print("검색 버튼 클릭 또는 결과 페이지 로드 중 오류 발생:", e)

finally:
    driver.quit()
```
